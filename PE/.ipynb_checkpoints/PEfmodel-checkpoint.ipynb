```python
X = dataset.drop(['legitimate','Name','md5'],axis=1).values
y = dataset['legitimate'].values 
```

```python
import os
import pandas
import pickle
import pefile
import sklearn.ensemble as ek
from sklearn.feature_selection import SelectFromModel
import joblib
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn import svm
import sklearn.metrics as metrics
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectFromModel
import seaborn as sns
import numpy
import matplotlib.pyplot as plt
```

```python
# important feature selection

extratrees = ExtraTreesClassifier().fit(X, y)
selector = SelectFromModel(extratrees, prefit=True)
X_new = selector.transform(X)                           
```

```python
# Plotting feature importances for all features in X in descending order

plt.figure(figsize=(14, 8))  

sorted_indices = numpy.argsort(extratrees.feature_importances_)[::-1]
sorted_importances = extratrees.feature_importances_[sorted_indices]
sorted_features = [dataset.columns[2 + i] for i in sorted_indices]

plt.bar(range(X.shape[1]), sorted_importances, align="center")
plt.xticks(range(X.shape[1]), sorted_features, rotation=90, ha='right')  
plt.xlabel("Feature")
plt.ylabel("Importance")
plt.title("Feature Importances (Descending Order)")
plt.tight_layout()  
plt.show()
```

```python
# Print important features with importances
n_features = X_new.shape[1]
features = [(dataset.columns[2+i], extratrees.feature_importances_[i]) for i in numpy.argsort(extratrees.feature_importances_)[::-1][:n_features]]

print("Selected Features:")
for f, (feature, importance) in enumerate(features):
    print("%d. feature %s (%f)" % (f + 1, feature, importance))
```

```python
# Compute the correlation matrix using the selected features
corr_matrix = pd.DataFrame(X_new, columns=features).corr()

# Plot the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title("Correlation Heatmap of Selected Features")
plt.show()
```

```python
# splitting the data (70:30)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.30, stratify = y)
```

```python
# finding the best classifier
!pip install tpot
from tpot import TPOTClassifier
tpot = TPOTClassifier(
        generations=5,
        population_size=20,
        verbosity=2,
        random_state=42,
        config_dict='TPOT sparse',
        scoring='accuracy',
        cv=5
)
```

```python
tpot.fit(X_new,y)
```

```python
best_pipeline = tpot.fitted_pipeline_
```

```python
from sklearn.metrics import accuracy_score, balanced_accuracy_score, precision_score, recall_score, f1_score

# Assuming you have a trained model 'model'
y_pred = best_pipeline.predict(X_test)

# Calculate metrics
accuracy = accuracy_score(y_test, y_pred)
balanced_accuracy = balanced_accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Print the results
print(f"Accuracy: {accuracy:.4f}")
print(f"Balanced Accuracy: {balanced_accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
```
